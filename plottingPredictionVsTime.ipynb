```python
%run singleRepoStats.py
import datetime
import pandas as pd
import featureMakers
```

```python
repoString = 'copitux/python-github3'
z= getRepoWeeklyData(repoString)

weeklyTotal = z.pivot(index='week_start', columns='author_login', values='commits_num').sum(axis=1)
weeklyTotalResampled = featureMakers.resampleToDays(weeklyTotal)

tref = datetime.date.today()

probAlive = getPredictedProbAlive(z, tref =tref)
```

```python
probAlive
```

```python
earliestCommit = z['week_start'].min()

aWeekAfterFirstCommit = earliestCommit + datetime.timedelta(7)

times = pd.date_range(start=aWeekAfterFirstCommit, 
                      freq='W', end=datetime.date.today())
times = [x.to_datetime().date() for x in times]
```

```python
times[:2]
```

```python
probs = [getPredictedProbAlive(z, t) for t in times]
```

```python
probTrace = pd.Series(data=probs, index=pd.to_datetime(times))
```

```python
weeklyTotalResampled.plot()
```

```python
probTraceResampled = probTrace.reindex(weeklyTotalResampled.index, method='ffill', limit=1)
```

```python
probTraceResampled = probTraceResampled.interpolate()
```

```python
probTraceResampled.plot()
```

```python
commitsAndProb = pd.DataFrame({'commits':weeklyTotalResampled, 'prob':probTraceResampled})
```
