```python
query = '''
    SELECT 
        id, full_name, name, owner_login, fork, created_at 
    FROM Repos
    '''
repos = pd.io.sql.read_sql(query, con)
```

```python
repos.head()
```

```python
pd.concat(g for _, g in repos.groupby("full_name") if len(g) > 1)
```

```python
query = '''
SELECT
    full_name
FROM
    Repos
GROUP BY
    full_name
HAVING
    COUNT(full_name) > 1
'''
pd.io.sql.read_sql(query, con)
```

```python
query = '''
SELECT * FROM Repos WHERE full_name IN (
    SELECT full_name
    FROM Repos
    GROUP BY full_name
    HAVING COUNT(full_name) > 1
)
'''
duplicatedRepos = pd.io.sql.read_sql(query, con)
```
