# Discovered that the duplicated repositories are just that: complete duplicates

It looks like repos can move up and down in the rankings during search. Totally reasonable. I will need to explicitly check if a repo is already in the database before adding it.

```python
findFirstDuplicateQuery = '''
SELECT r1.id
    FROM Repos r1 JOIN Repos r2
    ON (r1.full_name = r2.full_name AND r1.id < r2.id)
'''
pd.io.sql.read_sql(findFirstDuplicateQuery, con)
```

```python
query = "select full_name, downloaded_on from repos where id IN (%s)" % findFirstDuplicateQuery
pd.io.sql.read_sql(query, con)
```

```python
makeDuplicateIdTables = '''
CREATE TEMPORARY TABLE dupRepos AS 
SELECT r1.id AS id
    FROM Repos r1 JOIN Repos r2
    ON (r1.full_name = r2.full_name AND r1.id < r2.id)
'''
c.execute(makeDuplicateIdTables)
con.commit()
```

```python
query = "SELECT Repos.* FROM Repos, dupRepos WHERE Repos.id = dupRepos.id"
pd.io.sql.read_sql(query, con)
```

```python
dropDuplicateIdTables = "DROP TABLE dupRepos"
c.execute(dropDuplicateIdTables)
con.commit()
```
