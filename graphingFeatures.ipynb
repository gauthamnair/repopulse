```python
import pandas as pd
from ggplot import *
```

```python
byRepo = pd.read_csv('data/intermediate/byRepo.csv')
del byRepo['Unnamed: 0']
byRepo.shape
```

```python
byRepo['aliveOrDead'] = np.where(byRepo['futureCommits_num'] > 0,
                                 'alive',
                                 'dead')
```

```python
byRepo.head()
```

```python
byRepo['aliveOrDead'].value_counts()
```

```python
q = ggplot(byRepo, aes(x='daysSinceLastCommit'))
q += geom_bar(binwidth=14)
q += ylab('number of repos')
q += facet_wrap('aliveOrDead')
print q
```

```python
byRepo['log10(PastCommits + 1)'] = np.log10(byRepo['pastCommits_num']+1)
q = ggplot(byRepo, aes(x='log10(PastCommits + 1)'))
q += geom_bar()
q += ylab('number of repos')
q += facet_wrap('aliveOrDead')
q += xlim(0,4)
print q
```

```python
q = ggplot(byRepo, aes(x='daysSinceFirstCommit', 
                       y='daysSinceLastCommit'))
q += geom_point()
q += facet_wrap('aliveOrDead')
print q
```

```python
dfX = byRepo[['daysSinceLastCommit', 'pastCommits_num', 'daysSinceFirstCommit']]
```

```python
x = dfX.values
```

```python
y = np.where(byRepo['aliveOrDead']=='alive',1,0)
```

```python
y.shape
```

```python
x.shape
```

```python
full_name = byRepo['repo_full_name'].values
```

```python
full_name
```

```python
x
```

```python
predictors = x[:,[0]]
predictors.shape
```

```python
y.shape
```

```python
from sklearn import linear_model
learner = linear_model.LogisticRegression()
```

```python
learner.fit(predictors, y)
```

```python
predictions = learner.predict(predictors)
```

```python
predictions
```

```python
y
```

```python
from sklearn.metrics import classification_report
```

```python
print classification_report(y_true=y, y_pred=predictions)
```

```python
byRepo['prediction'] = np.where(predictions,'alive','dead')
```

```python
q = ggplot(byRepo, aes(x='daysSinceLastCommit', fill='prediction'))
q += geom_bar(binwidth=14)
q += ylab('number of repos')
q += facet_wrap('aliveOrDead')
print q
```

```python
toPlot = np.arange(0,365)
toPlot = toPlot.reshape((len(toPlot),1))
```

```python
probs = learner.predict_proba(toPlot)
```

```python
probAlive = pd.Series(data=probs[:,1],index=toPlot)
probAlive.index.name = 'daysSinceLastCommit'
```

```python
probAlive.plot()
```
