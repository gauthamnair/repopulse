```python
def plotROC(report):
    (fpr, tpr) = report.roc()
    plot(fpr, tpr, label='ROC curve (area = %0.2f)' % report.auc())
    plot([0, 1], [0, 1], 'k--')
    xlim([0.0, 1.0])
    ylim([0.0, 1.0])
    xlabel('False Positive Rate')
    ylabel('True Positive Rate')
    legend(loc="lower right")
```

```python
def evaluatePredictors(predictorsMaker, baseLearner=None):
    if baseLearner==None:
        baseLearner = linear_model.LogisticRegression()
    learner = sklearn.pipeline.Pipeline(
        [
        ('scaler', sklearn.preprocessing.StandardScaler()),
        ('model', baseLearner)]) 

    (X, colNames) = predictorsMaker.makePredictors(byRepo)

    evaluateModel = makeModelEvaluator(X, y)
    report = evaluateModel(learner)

#     print report.classification_report()
#     print "auc:", report.auc()

    learner.fit(X, y)
    insampleScore = learner.predict_proba(X)[:,1]
    (fpr, tpr, _) = sklearn.metrics.roc_curve(y, insampleScore)
    insampleAUC = sklearn.metrics.auc(fpr, tpr)
    print 'insample AUC:' , insampleAUC
    print 'crossval AUC:' , report.auc()
    
    return (learner, report, colNames)

def plotLogistic(learner, report, colNames):
    logistic = learner.named_steps['model']
    coefs = pd.Series(data=np.concatenate( (logistic.coef_[0], logistic.intercept_) ), 
        index = colNames + ['intercept'])
    
    subplot(1,2,1)
    coefs.plot(kind='barh')
    plt.title('AUC: '+ str(report.auc()))
    subplot(1,2,2)
    plotROC(report)
```
