```python
from sklearn import linear_model
import sklearn.cross_validation
import sklearn.metrics
import sklearn.ensemble
import sklearn.preprocessing
import sklearn.pipeline
import modelTools
import trainingData
import numpy as np
import pandas as pd
```

```python
byRepo = trainingData.load()
modelTools.addAliveOrDeadColumn(byRepo)
y = modelTools.makeTarget(byRepo)
allColumns = modelTools._predColumns
```

```python
byRepo['aliveOrDead'].value_counts()
```

```python
def evaluatePredictors(predColumns = allColumns):
	learner = sklearn.pipeline.Pipeline(
		[('scaler', sklearn.preprocessing.StandardScaler()),
		('logistic', linear_model.LogisticRegression())]) 

	(X, colNames) = modelTools.makePredictors(byRepo, predColumns = predColumns)

	evaluateModel = modelTools.makeModelEvaluator(X, y)
	report = evaluateModel(learner)

	print report.classification_report()
	print "auc:", report.auc()

	learner.fit(X, y)
	return (learner, colNames, report)
```

```python
def plotROC(report):
    clf()
    (fpr, tpr) = report.roc()
    plot(fpr, tpr, label='ROC curve (area = %0.2f)' % report.auc())
    plot([0, 1], [0, 1], 'k--')
    xlim([0.0, 1.0])
    ylim([0.0, 1.0])
    xlabel('False Positive Rate')
    ylabel('True Positive Rate')
    legend(loc="lower right")
```

```python
def effectOfExcludingColumns(toExclude = None, includeOnly = None):
    if toExclude != None:
        print "excluding ", toExclude, ":"
        predColumns = [p for p in allColumns if p not in toExclude]
    else:
        print "including only ", includeOnly, ":"
        predColumns = includeOnly
    (learner, colNames, report) = evaluatePredictors(predColumns=predColumns)
    logistic = learner.named_steps['logistic']
    coefs = pd.Series(data=np.concatenate( (logistic.coef_[0], logistic.intercept_) ), 
        index = colNames + ['intercept'])
    coefs.plot(kind='barh')
    plt.title('AUC: '+ str(report.auc()))
    figure()
    plotROC(report)
    return (coefs, report)
```
