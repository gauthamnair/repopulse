# Purely DateTimeIndex approach

```python
s = pd.Series(data=[5, 3], index=pd.to_datetime(['2013-04-02 10:00:00', '2013-04-04 10:00:00']))
s
```

```python
s.index
```

```python
s.reindex(index=pd.date_range('2013-03-31', periods=8, freq='D'), 
          method='ffill', 
          limit=2).fillna(0)
```

```python
dailySince2010=pd.date_range(start='2010-01-01', end='2014-01-01')
```

```python
daysInWeek = 7
x = hadleyPlyr['commits_num'].reindex(
    index=dailySince2010,
    method='ffill',
    limit=daysInWeek)
x = x/daysInWeek
x = x.fillna(0)
x.plot()
```
