# resampling to daily activity

```python
daysInWeek = 7
upToTwoYearsAgo = pd.date_range(
    end=datetime.date.today(), periods= 365 * 2)

def dailyRangeSinceBeginning(timeseries):
    return pd.date_range(start=timeseries.index[0], end=datetime.date.today())

def resampleToDays(weeklyCommits, daterange=None):
    if daterange == None:
        daterange = dailyRangeSinceBeginning(weeklyCommits)
    x = weeklyCommits.reindex(
        index=daterange,
        method='ffill',
        limit=daysInWeek)
    x = x/daysInWeek
    x = x.fillna(0)
    return x
```

```python
z = resampleToDays(hadleyPlyr['commits_num'])
z.plot()
```

```python
z.ix['2013-01-01':'2014-01-01'].plot()
```

# function to sum over a period into the future

```python
def totalCommitsInPeriodStarting(commitsSeries, tref = None, daysIntoFuture=6*30):
    if tref == None:
        tref = datetime.date.today() - datetime.timedelta(daysIntoFuture + 1)
    return commitsSeries.ix[tref:(tref + datetime.timedelta(daysIntoFuture))].sum()
```

```python
totalCommitsInPeriodStarting(hadleyPlyr['commits_num'], daysIntoFuture=365*10).sum()
```

```python
grouped = contrib.groupby(['repo_full_name', 'author_login'], as_index=False)['commits_num']
target = grouped.aggregate(totalCommitsInPeriodStarting)
target.head()
```

```python
byRepo = target.groupby('repo_full_name').sum()
byRepo.ix['hadley/plyr']
```

# function to get time since last commit

```python
weeklyCommitsHadleyPlyr = hadleyPlyr['commits_num']
weeklyCommitsHadleyPlyr.head()
```

```python
dailyCommitsHadleyPlyr = resampleToDays(weeklyCommitsHadleyPlyr)
dailyCommitsHadleyPlyr[:15]
```

```python
# tref = datetime.date(2012, 1, 1)
tref = datetime.date.today()
subsetted = weeklyCommitsHadleyPlyr[weeklyCommitsHadleyPlyr>0].ix[:tref]
subsetted.tail()
```

```python
subsetted.index[-1]
```

```python
def mostRecentNonZero(timeseries, tref):
    subsetted = timeseries[timeseries>0].ix[:tref]
    return subsetted.index[-1]
```

```python
mostRecentNonZero(weeklyCommitsHadleyPlyr, tref)
```

```python
t = mostRecentNonZero(dailyCommitsHadleyPlyr, tref)
t
```

```python
type(tref)
```

```python
t.to_datetime().date()
```

```python
deltaT = tref - t.to_datetime().date()
```

```python
deltaT.days
```

```python
def daysSinceLastNonZero(timeseries, tref):
    t = mostRecentNonZero(timeseries, tref)
    deltaT = tref - t.to_datetime().date()
    return deltaT.days
```

```python
daysSinceLastNonZero(weeklyCommitsHadleyPlyr, tref)
```

```python
daysSinceLastNonZero(dailyCommitsHadleyPlyr, tref)
```
