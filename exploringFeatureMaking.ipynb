```python
daysInWeek = 7
upToTwoYearsAgo = pd.date_range(
    end=datetime.date.today(), periods= 365 * 2)

def resampleToDays(weeklyCommits, daterange=upToTwoYearsAgo):
    x = weeklyCommits.reindex(
        index=daterange,
        method='ffill',
        limit=daysInWeek)
    x = x/daysInWeek
    x = x.fillna(0)
    return x
```

```python
z = resampleToDays(hadleyPlyr['commits_num'])
z.plot()
```

```python
z.ix['2013-01-01':'2014-01-01'].plot()
```

```python
def totalCommitsInPeriodStarting(commitsSeries, tref = None, daysIntoFuture=6*30):
    if tref == None:
        tref = datetime.date.today() - datetime.timedelta(daysIntoFuture + 1)
    return commitsSeries.ix[tref:(tref + datetime.timedelta(daysIntoFuture))].sum()
```

```python
totalCommitsInPeriodStarting(hadleyPlyr['commits_num'], daysIntoFuture=365*10).sum()
```

```python
grouped = contrib.groupby(['repo_full_name', 'author_login'])['commits_num']
grouped.apply(totalCommitsInPeriodStarting)
```
